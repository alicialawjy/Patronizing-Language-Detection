```python
import pandas as pd
```

```python
from sklearn.metrics import classification_report
```

```python
file_name = "eval_results.csv"
```

```python
df = pd.read_csv(file_name,index_col=0)
```

```python
df
```

```python
pred = df['pred']
actual = df['actual']
```

```python
#print(classification_report(actual, pred))
```

```python
predictions = []
for ls in pred:
    ls = ls.replace("[", "")
    ls = ls.replace("]", "")
    ls_split = ls.split(" ")
    predictions.extend(ls_split)
    #print(the_list)
    #print(type(the_list))
    #predictions.extend(i)
```

```python
actual_list = []
for ls_a in actual:
    ls_a = ls_a.replace("[", "")
    ls_a = ls_a.replace("]", "")
    ls_a_split = ls_a.split(" ")
    actual_list.extend(ls_a_split)
    #print(the_list)
    #print(type(the_list))
    #predictions.extend(i)
```

```python
print(classification_report(predictions, actual_list))
```

```python
from sklearn.metrics import f1_score
f1_score(actual_list, predictions, average='macro')
```
